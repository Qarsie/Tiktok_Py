```python
import pandas as pd
import os

# Define the data folder path
data_folder = 'data'

# Get all CSV files in the data folder
csv_files = [os.path.join(data_folder, file) for file in os.listdir(data_folder) if file.endswith('.csv')]
print(f"Found {len(csv_files)} CSV files in data folder:")
for file in csv_files:
    print(f"- {file}")

# Merge all CSV files into a single DataFrame
df = pd.concat([pd.read_csv(file) for file in csv_files], ignore_index=True)

```

```python
print(df.columns)
```

```python
import os

# Create merged folder if it doesn't exist
merged_folder = os.path.join('data', 'merged')
os.makedirs(merged_folder, exist_ok=True)

# Save to CSV
output_csv = os.path.join(merged_folder, 'merged_data.csv')
df.to_csv(output_csv, index=False, encoding='utf-8')

# Save to Excel
output_excel = os.path.join(merged_folder, 'merged_data.xlsx')
df.to_excel(output_excel, index=False, engine='openpyxl')

print(f"Saved merged data to:\n- {output_csv}\n- {output_excel}")

```

```python
# Check for duplicates (optional)
duplicates = df[df.duplicated()]
print(duplicates)
```

```python
# Remove duplicates
df.drop_duplicates(subset="video_id", inplace=True)
```

```python
# Save deduplicated data to CSV
output_dedup = os.path.join(merged_folder, 'merged_data_deduplicated.csv')
df.to_csv(output_dedup, index=False, encoding='utf-8')
print(f"Saved deduplicated data to:\n- {output_dedup}")
```

```python
# Fill missing hashtags with empty list as string
df["hashtags"] = df["hashtags"].fillna("[]")

# Drop rows missing essential metrics
df.dropna(subset=["description", "likes", "comments", "shares", "views"], inplace=True)

# Convert timestamp fields
df["posted_time"] = pd.to_datetime(df["posted_time"])
df["fetch_time"] = pd.to_datetime(df["fetch_time"])
```

```python
df["engagement_rate"] = df["total_engagement"] / df["followers"]
# Engagement per view
df["engagement_rate"] = df["total_engagement"] / df["views"]
df["engagement_percentile"] = df["total_engagement"].rank(pct=True)
```
